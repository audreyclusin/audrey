```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../python_decal_fall25/course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
print(data.shape)
```

```python
new_data = np.array(data)
magnitude = new_data[:, 0]
temperature = new_data[:, 1]
```

```python
minimum_mag = np.min(magnitude)
maximum_mag = np.max(magnitude)
minimum_temp = np.min(temperature)
maximum_temp = np.max(temperature)
print(f'Star #{np.argmin(magnitude)} has the smallest magnitude of {minimum_mag}')
print(f'Star #{np.argmax(magnitude)} has the largest magnitude of {maximum_mag}')
print(f'Star #{np.argmin(temperature)} has the lowest temperature of {minimum_temp}')
print(f'Star #{np.argmax(temperature)} has the highest temperature of {maximum_temp}')
```

```python
print(f'Star #{np.argmin(magnitude)} is the brightest')
print(f'Star #{np.argmax(magnitude)} is the dimmest')
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
```

```python
plt.scatter(log_temperature, log_magnitude, label = 'Stars')
plt.scatter(np.log(temperature[np.argmin(magnitude)]), np.log(np.min(magnitude)), c = 'black', label = 'Brightest')
plt.scatter(np.log(temperature[np.argmax(magnitude)]), np.log(np.max(magnitude)), c = 'purple', label = 'Dimmest')
plt.scatter(np.log(np.min(temperature)), np.log(magnitude[np.argmin(temperature)]), c = 'orange', label = 'Coldest')
plt.scatter(np.log(np.max(temperature)), np.log(magnitude[np.argmax(temperature)]), c = 'green', label = 'Hottest')
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title('Hertzsprung-Russel Diagram')
plt.xlabel('Magnitude (Log)')
plt.ylabel('Temperature (Log K)')
plt.legend()
```
