```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
```

```python
# make the figure
plt.figure(figsize = (8, 8))

#do the plotting
plt.plot(x, y)
plt.plot(x, y**3, c = 'yellowgreen')

#give me labels or give me death
plt.xlabel("glory")
plt.ylabel("death")
plt.title("title!")

#show it
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax [0,0].plot(x, y)
ax [0,1].plot(x, y**2)
ax [1,0].plot(x, y*2)
ax [1, 1].plot(x*2, y)
```
