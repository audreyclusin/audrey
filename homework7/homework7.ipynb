```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    y = A*np.sin(w*x)
    return y

x = np.linspace(0, 2*np.pi, 1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

fig, axs = plt.subplots(1, 3, figsize = (25, 5))

for a, w in zip(amplitude, frequency):
    axs[0].scatter(x, make_sine_wave(x, a, 1))
    axs[0].set_title("Varying Amplitude")
    axs[0].set_xlabel("X values")
    axs[0].set_ylabel("Y values")
    axs[1].scatter(x, make_sine_wave(x, 1, w))
    axs[1].set_title("Varying Freqency")
    axs[1].set_xlabel("X values")
    axs[1].set_ylabel("Y values")
    axs[2].scatter(x, make_sine_wave(x, a, w))
    axs[2].set_title("Varying Amplitude and Frequency")
    axs[2].set_xlabel("X values")
    axs[2].set_ylabel("Y values")
    
plt.show()
```
