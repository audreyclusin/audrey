```python
import numpy as np
print(f'Median latitude: {np.median(df['latitude'])}')
print(f'Median longitude: {np.median(df['longitude'])}')
```

```python
df['Absolute Value of Longitude'] = abs(df['longitude'])
print(df)
```

```python
print(df.shape)
#52 rows, 5 columns
```

```python
median_lat = np.median(df['latitude'])
median_long = np.median(df['longitude'])

lat_arr = np.array(df['latitude'])
long_arr = np.array(df['longitude'])

difference_in_lat = abs(lat_arr - median_lat)
difference_in_long = abs(long_arr - median_long)

states = np.array(df['state'])

closest_lat = states[np.argmin(difference_in_lat)]
farthest_lat = states[np.argmax(difference_in_lat)]
closest_long = states[np.argmin(difference_in_long)]
farthest_long = states[np.argmax(difference_in_long)]

closest_lat_difference = np.min(difference_in_lat)
farthest_lat_difference = np.max(difference_in_lat)
closest_long_difference = np.min(difference_in_long)
farthest_long_difference = np.max(difference_in_long)

print(f'The state closest in latitude to the state with the median latitude is {closest_lat}, with a difference of {(closest_lat_difference) :.2f}')
print(f'The state farthest in latitude to the state with the median latitude is {farthest_lat}, with a difference of {(farthest_lat_difference) :.2f}')
print(f'The state closest in longitude to the state with the median longitude is {closest_lat}, with a difference of {(closest_long_difference) :.2f}')
print(f'The state farthest in longitude to the state with the median longitude is {farthest_lat}, with a difference of {(farthest_long_difference) :.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('attention')
df = df.dropna()
```

```python
print(df.shape)
#60 rows, 5 columns
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print(list(df.columns))
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10,10))
ax[0,0].scatter(df['Unnamed: 0'][0:19], df['score'][0:19], label = 'Participant Scores')
ax[0,0].set_title('Scores While Attention is Divided')
ax[0,0].set_xlabel('Participant #')
ax[0,0].set_ylabel('Score')
ax[0,0].set_ylim(0, 10)
ax[0,0].legend(loc='upper right')
#Participant scores while attention is split

ax[0,1].scatter(df['Unnamed: 0'][40:59], df['score'][40:59], color = 'green')
ax[0,1].set_title('Scores While Focused')
ax[0,1].set_xlabel('Participant #')
ax[0,1].set_ylabel('Score')
ax[0,1].set_ylim(0, 10)
#Participant scores while attention is fully on the activity

hist_data = []
for i in range(len(df.columns)):
    if df.columns[i] == 'attention':
        hist_data.append(df[df.columns[i]])
    elif df.columns[i] == 'score':
        hist_data.append(df[df.columns[i]])
hist_data = pd.DataFrame(hist_data).T

sns.histplot(x = 'attention', y = 'score', data = hist_data, ax = ax[1, 0], color = 'red')
ax[1,0].set_title('Scores While Focused')
ax[1,0].set_xlabel('Attention Level')
ax[1,0].set_ylabel('Score')
ax[1,0].set_ylim(0, 10)
#How participants scored when their attention was focused vs divided

sns.boxplot(hist_data, ax = ax[1, 1], color = 'orange')
ax[1,1].set_title('Score Distribution')
ax[1,1].set_xlabel('Score')
ax[1,1].set_ylabel('Score Frequency')
ax[1,1].set_ylim(0, 10)
#Shows how well  most people scored 

plt.show()
```
