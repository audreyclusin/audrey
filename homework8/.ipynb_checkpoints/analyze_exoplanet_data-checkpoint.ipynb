```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exoplanet_data.csv', comment = '#')
```

```python
print(df.shape)
#567 rows, 92 columns
```

```python
print(df.columns)
#st_tefflim, sy_pnum, sy_kmagerr1, sy_gaiamagerr1, rastr
```

```python
'''
Planet Name: pl_name - name of the exoplanet
Host Name: hostname - name of the host star of the exoplanet
Discovery Method: discoverymethod - how the planet was discovered - transit method, microlensing, radial velocity, etc
Spectral Type: st_spectype - what elements are present in the planet
Equilibrium Temperature [K]: pl_eqt - what temperature allows the planet to be in thermal equilibrium - based on size of home star + orbital distance
Planet Mass or Mass*sin(i) [Jupiter Mass]: pl_msinij -  mass in terms of the mass of jupiter 
'''
```

```python
relevant_data = []
for i in range(len(df.columns)):
    if df.columns[i] == 'pl_name':
        relevant_data.append(df[df.columns[i]])
    elif df.columns[i] == 'hostname':
        relevant_data.append(df[df.columns[i]])
    elif df.columns[i] == 'discoverymethod':
        relevant_data.append(df[df.columns[i]])
    elif df.columns[i] == 'st_spectype':
        relevant_data.append(df[df.columns[i]])
    elif df.columns[i] == 'pl_eqt':
        relevant_data.append(df[df.columns[i]])
    elif df.columns[i] == 'pl_bmassj':
        relevant_data.append(df[df.columns[i]])
    
relevant_data = pd.DataFrame(relevant_data).T
relevant_data = relevant_data.dropna()
relevant_data
```

```python
print(relevant_data.shape)
#83 rows, 6 columns
```

```python
relevant_data.value_counts('discoverymethod')
'''
Transit: 68 - Observing when/how the host star's luminosity drops - indicates that an exoplanet is passing in front of the star
Radial velocity: 13 - observing how an object seems to undergo redshift or blueshift as it travels - indicates orbit
Imaging: 2 - looking for the visible planet 
'''
```

```python
plt.figure(figsize = (8, 6))

mask1 = (relevant_data['discoverymethod'] == 'Transit')
mask2 = (relevant_data['discoverymethod'] == 'Radial Velocity')
mask3 = (relevant_data['discoverymethod'] == 'Imaging')

plt.scatter(
    relevant_data['pl_bmassj'][mask1], relevant_data['pl_eqt'][mask1], c = 'green', label = 'Transit Discovered'
)
plt.scatter(
    relevant_data['pl_bmassj'][mask2], relevant_data['pl_eqt'][mask2], c = 'blue', label = 'Radial Velocity Discovered'
)
plt.scatter(
    relevant_data['pl_bmassj'][mask3], relevant_data['pl_eqt'][mask3], c = 'red', label = 'Imaging Discovered'
)
plt.title('Mass vs Equilibirium Temperature')
plt.xlabel("Mass (In terms of Jupiter's mass)")
plt.ylabel('Equilibirium Temperature')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8, 6))

mask1 = (relevant_data['discoverymethod'] == 'Transit')
mask2 = (relevant_data['discoverymethod'] == 'Radial Velocity')
mask3 = (relevant_data['discoverymethod'] == 'Imaging')

plt.scatter(
    (relevant_data['pl_bmassj'][mask1]), (relevant_data['pl_eqt'][mask1]), c = 'green', label = 'Transit Discovered',
)
plt.scatter(
    (relevant_data['pl_bmassj'][mask2]), (relevant_data['pl_eqt'][mask2]), c = 'blue', label = 'Radial Velocity Discovered'
)
plt.scatter(
    (relevant_data['pl_bmassj'][mask3]), (relevant_data['pl_eqt'][mask3]), c = 'red', label = 'Imaging Discovered'
)
plt.xscale('log')
plt.yscale('log')
plt.title('Mass vs Equilibirium Temperature')
plt.xlabel("Mass (In terms of Jupiter's mass)")
plt.ylabel('Equilibirium Temperature')
plt.legend()
plt.show()

#Planets must be large to be observed by imaging
```

```python
masses = np.array(relevant_data['pl_bmassj'])

more_than_5_j = []
for item in masses:
    if item > 5: 
        more_than_5_j.append(item)
        
more_than_5_j_temps = []
for i in range(3): 
    index_1 = np.where(relevant_data == more_than_5_j[i])[0]
    more_than_5_j_temps.append(relevant_data.iloc[int(index_1), 4])

fivem_and_cold_value = np.min(more_than_5_j_temps)
fivem_and_cold = int(np.where(relevant_data['pl_eqt'] == fivem_and_cold_value)[0])

planet = relevant_data.iloc[fivem_and_cold, 0]
print(f'The coldest planet with a mass larger than five Jupiter Masses is {planet} at index {fivem_and_cold}.')
print(f"The planet's equilibrium temperature is {relevant_data.iloc[fivem_and_cold, 4]}, it's mass is {relevant_data.iloc[fivem_and_cold, 3]}, and its star type is {relevant_data.iloc[fivem_and_cold, 5]}")
```

```python
plt.figure(figsize = (8, 6))

mask1 = (relevant_data['discoverymethod'] == 'Transit')
mask2 = (relevant_data['discoverymethod'] == 'Radial Velocity')
mask3 = (relevant_data['discoverymethod'] == 'Imaging')
mask4 = (relevant_data['pl_name'] == 'eps Ind A b')

plt.scatter(
    (relevant_data['pl_bmassj'][mask1]), (relevant_data['pl_eqt'][mask1]), c = 'green', label = 'Transit Discovered',
)
plt.scatter(
    (relevant_data['pl_bmassj'][mask2]), (relevant_data['pl_eqt'][mask2]), c = 'blue', label = 'Radial Velocity Discovered'
)
plt.scatter(
    (relevant_data['pl_bmassj'][mask3]), (relevant_data['pl_eqt'][mask3]), c = 'red', label = 'Imaging Discovered'
)
plt.scatter(
    (relevant_data['pl_bmassj'][mask4]), (relevant_data['pl_eqt'][mask4]), c = 'purple', s = 100, label = 'Coldest Planet > 5 Jupiters'
)

plt.xscale('log')
plt.yscale('log')
plt.title('Mass vs Equilibirium Temperature')
plt.xlabel("Mass (In terms of Jupiter's mass)")
plt.ylabel('Equilibirium Temperature')
plt.legend()
plt.show()
```
