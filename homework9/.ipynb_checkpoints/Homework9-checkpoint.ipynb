```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('../python_decal_fall25/course_assignments/homework9/noisy_curvefit_data.csv')
```

```python
print(df.shape)
#100 rows, 6 columns
```

```python
df.head()
```

```python
print(df.columns)
```

```python
plt.scatter(
    df['x'], df['y_noise_1'], c = 'gray', label = 'noise'
)
plt.title('X Values vs Y noise')
plt.xlabel('X Values')
plt.ylabel('Y noise')
plt.legend()
```

```python
def model(x, A, k):
    return A * np.exp(-k * x)

p0 = [1, 1]

popt, pcov = curve_fit(model, df['x'], df['y_noise_1'], p0=p0)

A_fit, k_fit = popt

print(f"Fit: y = {A_fit:.2f} * e^(-{k_fit:.2f} * x)")

residuals = df['y_noise_1'] - model(df['x'], A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(df['x']) - len(popt)
chi2_reduced = chi2 / dof
```

```python
plt.scatter(
    df['x'], df['y_noise_1'], c = 'gray', label = 'noise'
)

plt.plot(df['x'], (A_fit * np.e**(-k_fit * df['x'])), label = 'True Model', c = 'black', linestyle = '--'
        )
plt.plot(df['x'], model(df['x'], 5, 0.8))
plt.title(f'X Values vs Y noise (noise level = 1, chi squared = {chi2:.2f}')
plt.xlabel('X Values')
plt.ylabel('Y noise')
plt.legend()
```

```python
def fit_model(x, y):
    p0 = [1, 1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt
    residuals = y - model(x, A_fit, k_fit)
    chi2 = np.sum(residuals**2)
    dof = len(x) - len(popt)
    chi2_reduced = chi2 / dof
    return [A_fit, k_fit, chi2_reduced]
```

```python
fig, axs =  plt.subplots(2, 2, figsize = (12, 10))

#Noise Level 2
fit2 = fit_model(df['x'], df['y_noise_2'])
axs[0,0].scatter(df['x'], df['y_noise_2'], c = 'gray', label = 'noise')
axs[0,0].plot(df['x'], fit2[0] * np.e**(-fit2[1]*df['x']), c = 'red', label = 'fitted line')
axs[0,0].plot(df['x'], model(df['x'], 5, 0.8), c = 'black', label = 'True Fit', linestyle = '--')
axs[0,0].set_title(f'Noise vs Fitted Curve (noise level = 2, chi squared = {fit2[2]:.2f}')

#Noise Level 3
fit3 = fit_model(df['x'], df['y_noise_3'])
axs[0,1].scatter(df['x'], df['y_noise_3'], c = 'gray', label = 'noise')
axs[0,1].plot(df['x'], fit3[0] * np.e**(-fit3[1]*df['x']), c = 'red', label = 'fitted line')
axs[0,1].plot(df['x'], model(df['x'], 5, 0.8), c = 'black', label = 'True Fit', linestyle = '--')
axs[0,1].set_title(f'Noise vs Fitted Curve (noise level = 3, chi squared = {fit3[2]:.2f}')

#Noise Level 4
fit4 = fit_model(df['x'], df['y_noise_4'])
axs[1,0].scatter(df['x'], df['y_noise_4'], c = 'gray', label = 'noise')
axs[1,0].plot(df['x'], fit4[0] * np.e**(-fit4[1]*df['x']), c = 'red', label = 'fitted line')
axs[1,0].plot(df['x'], model(df['x'], 5, 0.8), c = 'black', label = 'True Fit', linestyle = '--')
axs[1,0].set_title(f'Noise vs Fitted Curve (noise level = 4, chi squared = {fit4[2]:.2f}')

#Noise Level 5
fit5 = fit_model(df['x'], df['y_noise_5'])
axs[1,1].scatter(df['x'], df['y_noise_5'], c = 'gray', label = 'noise')
axs[1,1].plot(df['x'], fit5[0] * np.e**(-fit5[1]*df['x']), c = 'red', label = 'fitted line')
axs[1,1].plot(df['x'], model(df['x'], 5, 0.8), c = 'black', label = 'True Fit', linestyle = '--')
axs[1,1].set_title(f'Noise vs Fitted Curve (noise level = 5, chi squared = {fit5[2]:.2f}')
```

```python
'''
Questions: 
1. As the noise level increases, the red lie becomes flatter and less similar to the black line
2. Reduced chi squared increases as you increase the noise level, because the generated curve fit line is less simialr to the true curve fit. This is because the data is less organized - noisier.
'''
```

```python
fit1 = fit_model(df['x'], df['y_noise_1'])
noise_level = [1, 2, 3, 4, 5]
A_fits = [fit1[0], fit2[0], fit3[0], fit4[0], fit5[0]]
k_fits = [fit1[1], fit2[1], fit3[1], fit4[1], fit5[1]]
chi2s = [fit1[2], fit2[2], fit3[2], fit4[2], fit5[2]]

data = [noise_level, A_fits, k_fits, chi2s]

noisy_parameters = pd.DataFrame(data, columns = ['Noise Level 1', 'Noise Level 2', 'Noise Level 3', 'Noise Level 4', 'Noise Level 5'])
noisy_parameters.index = ['Noise Level', 'A fit', 'k fit', 'Reduced Chi Square']

print(noisy_parameters)
```

```python
noisy_parameters.to_csv("noisy_parameters.csv", index=False)
```
